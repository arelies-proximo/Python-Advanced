# REGEX FILTERING

```python
df.loc[ df.Name.str.contains("Mega")]
```

```python
#Rows which do not contain mega
df.loc[ (~df.Name.str.contains("Mega")) & (df.HP > 150)]
```

```python
import re
```

```python
df.loc[ df["Type 1"].str.contains("Grass|Fire")]
```

```python
df.loc[ df["Type 1"].str.contains('Grass|Fire', regex=True)]
```

```python
#Name which starts with pi
df.loc[ df.Name.str.contains("pi[a-z]*")]
```

```python
df.loc[ df.Name.str.contains("^pi[a-z]*")]
```

```python
df.loc[ df.Name.str.contains("^pi[a-z]*", regex=True) ]
```

```python
df.loc[ df.Name.str.contains("^pi[a-z]*", flags=re.I, regex=True) ]
```

**flags=re.I**


For ignoring Case in regex
