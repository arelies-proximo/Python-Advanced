## **Method Resolution Order (MRO) in Python**





Method Resolution Order (MRO) in Python is the order in which the Python interpreter looks for methods and attributes in a class hierarchy when they are accessed via an object. MRO plays a crucial role in multiple inheritance scenarios, where a class inherits from multiple parent classes.


The MRO is determined using the C3 linearization algorithm, also known as C3 superclass linearization. It ensures that the methods and attributes of parent classes are resolved in a consistent and predictable order.




Here's how the MRO works:

1. When you call a method or access an attribute on an object, Python first looks for it in the current class.
2. If it's not found in the current class, Python looks in the parent classes in a specific order determined by the C3 linearization algorithm.
3. The MRO order is typically determined by the method resolution order of the base classes, following a depth-first, left-to-right approach.

```python
#EXAMPLE

class A:
    def show(self):
        print("A")

class B(A):
    def show(self):
        print("B")

class C(A):
    def show(self):
        print("C")

class D(B, C):
    pass

obj = D()
obj.show()  # This will print "B" because B is the first class in the MRO of D.
print(D.mro())  # Prints the MRO for class D.
```

**In this example, the MRO for class D is [D, B, C, A], which means that when we call obj.show(), Python looks for the show method in D, then B, then C, and finally A. This follows the MRO order determined by the C3 linearization algorithm.**

#### Example of Inheritance

```python
class Polygon:
    # Initializing the number of sides
    def __init__(self, no_of_sides):
        self.n = no_of_sides
        self.sides = [0 for i in range(no_of_sides)]

    def inputSides(self):
        self.sides = [float(input("Enter side "+str(i+1)+" : ")) for i in range(self.n)]

    # method to display the length of each side of the polygon
    def dispSides(self):
        for i in range(self.n):
            print("Side",i+1,"is",self.sides[i])



class Triangle(Polygon):
    # Initializing the number of sides of the triangle to 3 by
    # calling the __init__ method of the Polygon class
    def __init__(self):
        Polygon.__init__(self,3)

    def findArea(self):
        a, b, c = self.sides

        # calculate the semi-perimeter
        s = (a + b + c) / 2

        # Using Heron's formula to calculate the area of the triangle
        area = (s*(s-a)*(s-b)*(s-c)) ** 0.5
        print('The area of the triangle is %0.2f' %area)


t = Triangle()

t.inputSides()

# Displaying the sides of the triangle
t.dispSides()
```

## **ENCAPSULATION**


**BINDING DATA MEMBERS N FUNCTIONS**

*Achieved through **DATA HIDING** using private and protection*


In Python, underscores _ and double underscores __ are used to indicate encapsulation and control the visibility and accessibility of class members (attributes and methods).

**Private variables** are declared with a **double underscore (__)** at the beginning of the name. They can **only be accessed within the class** in which they are declared.

**Protected variables** are declared with a **single underscore (_)** at the beginning of the name.
They can be accessed within the class in which they are declared and in any **subclasses of that class**.



```python
class MyClass:
    def __init__(self):
        self.__private_var = 10

    def __private_method(self):
         print("This is a private method.")

    def print_private(self):
      print(self.__private_var)

    def access_private_methods(self):
      self.__private_method()

obj = MyClass()

obj.print_private()
obj.access_private_methods()

```

### Getter and Setter are used as to reduce chances of inconsistency in the data members

```python
class Circle:
    def __init__(self, radius):
        self._radius = radius  # Private attribute with a single leading underscore

    # Getter method to get the radius
    def get_radius(self):
        return self._radius

    # Setter method to set the radius
    def set_radius(self, new_radius):
        if new_radius >= 0:
            self._radius = new_radius
        else:
            print("Radius cannot be negative.")
```

```python
circle = Circle(5)
print("Radius:", circle.get_radius())

circle.set_radius(7)
print("New Radius:", circle.get_radius())

circle.set_radius(-2)
print("Radius after invalid update:", circle.get_radius())
```

## **POLYMORPHISM**




Existence in more than one form.

```python
class SQR:
  def area(self,a):
    print("The area of Sqaure is :", a*a)

  def area(self, a, b):
    print("Area of Rectange is: ", a*b)


obj = SQR()

obj.area(25)

obj.area(50,10)
```

The issue with your code is that you're trying to define multiple methods with the same name (area) in the same class. In Python, when you define multiple methods with the same name in a class, the **last definition overwrites any previous ones**. This is known as **method overloading**.

**Achieving Polymorphism through IF ELSE**

```python
class SQR:
    def area(self, a, b=None):
        if b is None:
            print("The area of Square is:", a * a)
        else:
            print("Area of Rectangle is:", a * b)

obj = SQR()

obj.area(25)        # Computes the area of a square
obj.area(50, 10)    # Computes the area of a rectangle
```

### **super()._ _ init_ _(a)**

*Calling the super constructor/initializer*

```python
class SQUARE:
    def __init__(self, a):
        self.a = a

    def area(self):
        print(f"The area with side {self.a} is: {self.a * self.a}")

class RECT(SQUARE):
    def __init__(self, a, b):
        super().__init__(a)  # Call the superclass (SQUARE) __init__ method to initialize 'a'
        self.b = b

    def area(self):
        print(f"The area with sides {self.a} and {self.b} is {self.a * self.b}")

# Create instances of SQUARE and RECT
square = SQUARE(5)
rectangle = RECT(4, 6)

# Calculate and print the areas
square.area()
rectangle.area()
```

Chaning the way different user defined objects work wrt operators like +, -, print() etc.

```python
class Point:
    def __init__(self, x=0, y=0):
        self.x = x
        self.y = y

    def __str__(self):
        return "({0},{1})".format(self.x, self.y)

    def __add__(self, other):
        x = self.x + other.x
        y = self.y + other.y
        return Point(x, y)


p1 = Point(1, 2)
p2 = Point(2, 3)
```
